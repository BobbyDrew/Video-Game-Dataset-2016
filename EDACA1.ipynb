### Detection of outliers using z-zscore method

```python
from scipy import stats
```

```python
z=stats.zscore(a["NA_Sales"])
z
```

### Positions of z-score

```python
as1=np.where(z>0)
as1
```

```python
a.drop(as1[0],inplace=True)
```

### Detection of outliers using IQR method

```python
Q1=np.percentile(a["NA_Sales"],25,interpolation='midpoint')
Q3=np.percentile(a["NA_Sales"],75,interpolation='midpoint')
IQR=Q3-Q1
print(Q1)
print(Q3)
print(IQR)
print("shape",a.shape)
```

```python
upper=a["NA_Sales"]>=(Q3+1.5*IQR)
print("upper: ",upper)
print(np.where(upper))
lower=a["NA_Sales"]<=(Q3-1.5*IQR)
print("lower: ",lower)
print(np.where(lower))
print("new shape",a.shape)
```
