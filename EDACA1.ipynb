# Attribute Information

# Name : Name of the video game
##### Platform : In which platform the game will be available
##### Year_of_Release : In which year the game was released
##### Genre : Shows which type of game is it
##### Publisher : Tells who is the publisher
##### NA_Sales : Shows the North Americs Sales of the data
##### EU_Sales : Shows the Europe Sales of the data
##### JP_Sales : Shows the Japan Sales of the data
##### Other_Sales : Shows the other Countries sales
##### Global_Sales : Shows the Global Sales of the data
##### Critic_Score : Shows the critic score of the data
##### Critic_Count : Shows the critic count of the data
##### User_Score : Shows the user score 
##### User_Count : Tells the user count
##### Developer : Shows the developer of the game
##### Rating : Shows the rating of the game

# Histplot

```python
a[["NA_Sales","EU_Sales","JP_Sales","Other_Sales"]].describe()
```

### From the above table we can see that the NA_Sales has the highest values
### Lets visualize the graph between these columns and find which has the largest skew 

```python
plt.figure(figsize=(20,14))
plt.subplot(231)
plt.title("NA_Sales")
sns.histplot(a["NA_Sales"])
plt.subplot(232)
plt.title("EU_Sales")
sns.histplot(a["EU_Sales"])
plt.figure(figsize=(20,14))
plt.subplot(231)
plt.title("JP_Sales")
sns.histplot(a["JP_Sales"])
plt.subplot(232)
plt.title("Other_Sales")
sns.histplot(a["Other_Sales"])
plt.show()
```

### We can see here that the NA_Sales has the highest skew point 
### Let's check for Global_Sales

```python
a["Global_Sales"].describe()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(a["Global_Sales"])
plt.show()
```

### From above we can conclude that Global_Sales has the highest skew value compared to other variables

### Let's check the histogram for NA_Sales between 10-20 million in USD

### Finding the ouliers of the variables using Boxplot

```python
s=["NA_Sales","EU_Sales","JP_Sales","Other_Sales"]
plt.figure(figsize=(12,6))
a[s].boxplot()
plt.show()
```

### Now let's see for Global_Sales

```python
plt.figure(figsize=(12,6))
sns.boxplot(a["Global_Sales"])
plt.show()
```

### These are the outliers of the Global_Sales

```python
# Shows the maximum value of the Global_Sales
x=a["Global_Sales"].max()
print("The maximum value of the Global_Sales is {}".format(x))
```

### Let's see the KDE plot using Distplot representing Histogram graph as False

```python
plt.figure(figsize=(20,14))
plt.subplot(231)
plt.title("NA_Sales")
sns.distplot(a["NA_Sales"],hist=False)
plt.subplot(232)
plt.title("EU_Sales")
sns.distplot(a["EU_Sales"],hist=False)
plt.figure(figsize=(20,14))
plt.subplot(231)
plt.title("JP_Sales")
sns.distplot(a["JP_Sales"],hist=False)
plt.subplot(232)
plt.title("Other_Sales")
sns.distplot(a["Other_Sales"],hist=False)
plt.show()
```

### Now lets see KDE plot for Global_Sales

```python
sns.distplot(a["Global_Sales"],hist=False)
plt.figure(figsize=(20,14))
plt.show()
```
